```python

website_url = requests.get('https://en.wikipedia.org/wiki/Subprefectures_of_Hokkaido').text

soup = BeautifulSoup(website_url,'lxml')
print(soup.prettify())
```

```python
table = soup.find('table',{'class':'wikitable'})
table
```

```python
df_data=pd.read_html(str(table))
# convert list to dataframe
df_data =pd.DataFrame(df_data[0])
df_data.columns = ['Subprefecture index','Subprefecture',
                        'Japanese',
                        'Capital',
                        'Largest Municipality',
                        'Pop (2009)',
                        'Area',
                        'Cities', 'Towns', 'Villages']
df_data
```

```python

df_data = df_data.drop(['Subprefecture index', 'Villages'], axis=1)
df_data
```

```python

df_data['Subprefecture']=df_data['Subprefecture'].str.replace('↳', ' ')
df_data.head()
```
