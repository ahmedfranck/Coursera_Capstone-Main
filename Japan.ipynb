```python
from bs4 import BeautifulSoup

import numpy as np # library to handle data in a vectorized manner

import pandas as pd # library for data analsysis
import requests # library to handle requests
```

```python
html_data=requests.get('https://en.wikipedia.org/wiki/Subprefectures_of_Hokkaido').text
```

```python
soup=BeautifulSoup(html_data, 'lxml')
soup
```

```python
column_names = ['Subprefecture', 'Japan', 'Capital', 'LM', 'Pop', 'Area', 'M']
df = pd.DataFrame(columns=column_names)
df
```

```python
table = soup.find('table')
row_num = 0
for row in table.find("tbody").find_all("tr"):
    
    #print(row)
    col = row.find_all("b")
    #print(col)
    if (col != []):
        subprefecture = col[0].text
        japanese = col[1].text
        capital = col[2].text
        largestMunicipality = col[3].text
        pop = col[4].text
        area = col[5].text
        mun = col[6].text
        
        df = df.append({'Subprefecture': subprefecture,
                        'Japanese':japanesey,
                        'Capital': capital,
                        'LargestMunicipality': largestMunicipality,
                        'Pop': pop,
                        'Area': area,
                        'Mun': mun
                       }, ignore_index=True)
```
